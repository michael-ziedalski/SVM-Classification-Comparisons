### Imports

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['figure.figsize'] = (6.0, 3.0)
matplotlib.rcParams['figure.dpi'] = 150

from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import sklearn.svm as svm

from mlxtend.plotting import plot_decision_regions
```

### Data

```python
data = datasets.load_iris()

## For 2 categories
two_cats_x = data['data'][:, (2,3)]
two_cats_y = (data['target'] == 2).astype(np.float64)

## For 3 categories
# x is same as before, only three_cats_y is different.
three_cats_x = two_cats_x
three_cats_y = data['target'].astype(np.float64)
```

## SVM classification of 2 and 3 different categories, respectively
### Linear boundaries

```python
## Fitting and predicting
svm_two_class = Pipeline([
    ('scalar', StandardScaler()),
    ('linear_two_class_svm', svm.LinearSVC(C=1, loss='hinge'))
])

fit_two_class = svm_two_class.fit(two_cats_x, two_cats_y)

## Random prediction gives either 0 or 1, for either group.
print(fit_two_class.predict([[2.5, 1.7]]))

## Plotting
plot_decision_regions(two_cats_x, two_cats_y.astype(np.int16), clf=fit_two_class, legend=2)
```

```python
## Fitting and predicting 3 categories
svm_three_class = Pipeline([
    ('scalar', StandardScaler()),
    ('linear_three_class_svm', svm.LinearSVC(C=1, loss='hinge'))
])

fit_three_class = svm_three_class.fit(three_cats_x, three_cats_y)

## Random prediction gives either 0 or 1, for either group.
fit_three_class.predict([[2.5, 1.7]])

## Plotting
plot_decision_regions(three_cats_x, three_cats_y.astype(np.int16), clf=fit_three_class, legend=3)
```

### Non-linear boundaries.
#### - Polynomial

```python
## Fitting and predicting
svm_two_class = Pipeline([
    ('scalar', StandardScaler()),
    ('non-linear_two_class_svm', svm.SVC(kernel='poly', degree=3, C=1))
])

non_linear_two_class = svm_two_class.fit(two_cats_x, two_cats_y)

## Random prediction gives either 0 or 1, for either group.
print(non_linear_two_class.predict([[2.5, 1.7]]))

## Plotting
plot_decision_regions(two_cats_x, two_cats_y.astype(np.int16), clf=non_linear_two_class, legend=2)
```

```python
## Fitting and predicting
svm_three_class = Pipeline([
    ('scalar', StandardScaler()),
    ('non-linear_three_class_svm', svm.SVC(kernel='poly', degree=3, C=1))
])

non_linear_three_class = svm_three_class.fit(three_cats_x, three_cats_y)

## Random prediction gives either 0 or 1, for either group.
print(non_linear_three_class.predict([[2.5, 1.7]]))

## Plotting
plot_decision_regions(three_cats_x, three_cats_y.astype(np.int16), clf=non_linear_three_class, legend=3)
```

#### - Radial ('rbf')

```python
## Fitting and predicting
svm_two_class = Pipeline([
    ('scalar', StandardScaler()),
    ('non-linear_two_class_svm', svm.SVC(kernel='rbf', C=1))
])

non_linear_two_class = svm_two_class.fit(two_cats_x, two_cats_y)

## Random prediction gives either 0 or 1, for either group.
print(non_linear_two_class.predict([[2.5, 1.7]]))

## Plotting
plot_decision_regions(two_cats_x, two_cats_y.astype(np.int16), clf=non_linear_two_class, legend=2)
```

```python
## Fitting and predicting 3 categories
svm_three_class = Pipeline([
    ('scalar', StandardScaler()),
    ('non-linear_two_class_svm', svm.SVC(kernel='rbf', C=1))
])

non_linear_three_class = svm_three_class.fit(three_cats_x, three_cats_y)

## Random prediction gives either 0 or 1, for either group.
print(non_linear_three_class.predict([[2.5, 1.7]]))

## Plotting
plot_decision_regions(three_cats_x, three_cats_y.astype(np.int16), clf=non_linear_three_class, legend=2)
```
